```python
# Importar bibliotecas
import pandas as pd
import numpy as np
```

```python
# Importando a base de dados
saude_sono = pd.read_csv('/content/1KBtJf20QHGyrNOTEGeq_saude_do_sono_estilo_vida.csv')
```

```python
# 1. Renomeando colunas
saude_sono = saude_sono.rename(columns={
    'ID': 'Identificador',
    'Pressão sanguíneaaaa': 'Pressão sanguínea',
    'Ocupação': 'Profissão',
    'Categoria BMI': 'Categoria IMC'})
saude_sono
```

```python
# 2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões?
profession_statistics = saude_sono.groupby('Profissão')['Duração do sono'].agg(['mean', 'median'])
profession_statistics['mode'] = saude_sono.groupby('Profissão')['Duração do sono'].apply(lambda x: x.mode().iloc[0])
profession_statistics
```

```python
# 3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?
  # Filtrar Engenheiros de Software:
software_engineers = saude_sono[saude_sono['Profissão'] == 'Eng. de Software']
  # Filtrar Engenheiros de Software Obesos
software_engineers_bmi = software_engineers[software_engineers['Categoria IMC'] == 'Obesidade']
  # Calcular a Porcentagem
software_engineers_bmi_percentage = (len(software_engineers_bmi) / len(software_engineers)) * 100
print(f"\n {software_engineers_bmi_percentage:.2f} % dos Engenheiros de Software são Obesos ")
```

```python
# 4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos?
  # Filtrar Profissões de Interesse e calcular media
profession_interest = saude_sono[(saude_sono['Profissão'] == 'Advogado(a)') | (saude_sono['Profissão'] == 'Representante de Vendas')]
average_sleep_hours = profession_interest.groupby('Profissão')['Duração do sono'].mean()

  # Filtrar Profissões Não de Interesse e calcular media
other_professions = saude_sono[~saude_sono['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
average_sleep_other = other_professions['Duração do sono'].mean()
print(f"\nMédia de duração do sono para outras profissões: {average_sleep_other:.2f} horas")

  # Comparar médias
media_interesse = average_sleep_hours.mean()
print(f"\nMédia de duração do sono para Advogado(a) e Representante de Vendas: {media_interesse:.2f} horas")

if media_interesse < average_sleep_other:
    print("\nAdvogado(a) e Representante de Vendas dormem menos em média.")
else:
    print("\nAdvogado(a) e Representante de Vendas dormem mais ou igual em média.")
```

```python
# 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono?
nurse = saude_sono[saude_sono['Profissão'] == 'Enfermeiro(a)']
doctor = saude_sono[saude_sono['Profissão'] == 'Médico(a)']
nurse_average_sleep = nurse['Duração do sono'].mean()
doctor_average_sleep = doctor['Duração do sono'].mean()
print(f"Média de duração do sono para enfermeiros: {nurse_average_sleep:.2f} horas")
print(f"Média de duração do sono para médicos: {doctor_average_sleep:.2f} horas")
if nurse_average_sleep < doctor_average_sleep:
    print("Enfermeiros dormem menos em média do que médicos.")
else:
    print("Médicos dormem menos em média do que enfermeiros.")
```

```python
# 7. Descubra qual a profissão menos frequente no conjunto.
profession_counts = saude_sono['Profissão'].value_counts()
least_frequent_profession = profession_counts.idxmin()
print(f"A profissão menos frequente é: {least_frequent_profession}")
```

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres?

  # Separar a coluna 'Pressão sangre' em duas novas colunas
saude_sono[['Pressão Sistólica', 'Pressão Diastólica']] = saude_sono['Pressão sanguínea'].str.split('/', expand=True)

  # Convertir as novas colunas em tipo numérico
saude_sono['Pressão Sistólica'] = pd.to_numeric(saude_sono['Pressão Sistólica'])
saude_sono['Pressão Diastólica'] = pd.to_numeric(saude_sono['Pressão Diastólica'])

  # Eliminar a coluna original 'Pressão sanguínea'
saude_sono = saude_sono.drop(columns=['Pressão sanguínea'])

  # Calcular a media
average_blood_pressure_systolic = saude_sono.groupby('Gênero')['Pressão Sistólica'].mean()
average_blood_pressure_diastolic = saude_sono.groupby('Gênero')['Pressão Diastólica'].mean()

  # Identificar o gênero com a maior pressão média
gender_with_highest_systolic = average_blood_pressure_systolic.idxmax()
gender_with_highest_diastolic = average_blood_pressure_diastolic.idxmax()

print(f"Gênero com a maior pressão sistólica média: {gender_with_highest_systolic}")
print(f"Gênero com a maior pressão diastólica média: {gender_with_highest_diastolic}")
```

```python
# 9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?
mode = saude_sono['Duração do sono'].mode().iloc[0]
sleep_8_hours = mode == 8

print(f"A moda da duração do sono é: {mode} horas")
print(f"É predominante entre os participantes dormir 8 horas por dia? {'Sim' if sleep_8_hours else 'Não'}")
```

```python
# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70?
above_70 = saude_sono[saude_sono['Frequência cardíaca'] > 70]
below_equal_70 = saude_sono[saude_sono['Frequência cardíaca'] <= 70]
average_steps_above_70 = above_70['Passos diários'].mean()
average_steps_below_equal_70 = below_equal_70['Passos diários'].mean()
print(f"Média de passos diários de pessoas com frequência cardíaca acima de 70: {average_steps_above_70:.2f} passos")
print(f"Média de passos diários de pessoas com frequência cardíaca abaixo ou igual a 70: {average_steps_below_equal_70:.2f} passos")
if average_steps_above_70 > average_steps_below_equal_70:
    print("Pessoas com frequência cardíaca acima de 70 dão mais passos.")
else:
    print("Pessoas com frequência cardíaca abaixo ou igual a 70 dão mais passos.")
```
